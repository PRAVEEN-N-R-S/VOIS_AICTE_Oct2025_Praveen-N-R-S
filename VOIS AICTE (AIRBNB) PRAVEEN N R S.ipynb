```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Uploading Dataset
df = pd.read_excel("1730285881-Airbnb_Open_Data.xlsx")
```

```python
#Basic Info 
print("Shape:", df.shape)
print(df.info())
display(df.head())

#summary stats
display(df.describe(include="all"))
```

```python
# Calculate the mean price for each neighbourhood (country-wise) group
average_price_by_group = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)

print("--- Analysis: Average Price by Neighbourhood Group ---")
print(average_price_by_group.round(2))
print("\nDescriptive Statistics for Price (All Listings):")
print(df['price'].describe().round(2))
```

```python
#VISUALIZATIONS

# Setting plot style
sns.set(style="whitegrid", palette="muted", font_scale=1.1)



#1. Price Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=100, kde=True)
plt.xlim(0, 500)   # cap outliers
plt.title("Distribution of Listing Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.show()

# Price distribution (clip extreme outliers at 99th percentile for readability)
airbnb_df = df.copy() #dummy data/copy data
price_clip = airbnb_df['price'].clip(upper=airbnb_df['price'].quantile(0.99))
plt.figure(figsize=(8,4))
sns.histplot(price_clip.dropna(), bins=40, kde=True, color='#2a9d8f')
plt.title('Price distribution (clipped at 99th percentile)')
plt.xlabel('Price')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#  2. Room Type Popularity 
plt.figure(figsize=(8,5))
sns.countplot(x="room type", data=df, order=df["room type"].value_counts().index)
plt.title("Room Type Popularity")
plt.ylabel("Number of Listings")
plt.xlabel("Room Type")
plt.show()
```

```python
# 3. Availability (days per year) 
plt.figure(figsize=(10,5))
sns.histplot(df["availability 365"], bins=50)
plt.title("Availability (days per year)")
plt.xlabel("Days available")
plt.ylabel("Count")
plt.show()
```

```python
# 4. Reviews Distribution 
plt.figure(figsize=(10,5))
sns.histplot(df['number of reviews'], bins=50)
plt.xlim(0, 200)
plt.title("Distribution of Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Average Price by Neighbourhood (country-wise) Group 

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
# Create a bar plot for the average price by group
sns.barplot(
    x=average_price_by_group.index,
    y=average_price_by_group.values,
    palette="viridis"
)

# Adding titles and labels
plt.title('Average Airbnb Price by country-wise', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price (USD $)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 6. Host Listings Count
plt.figure(figsize=(10,5))
sns.histplot(df["calculated host listings count"], bins=50)
plt.title("Distribution of Host Listings Count")
plt.xlabel("Number of Listings per Host")
plt.ylabel("Count")
plt.show()
```

```python
# 7. Map of Listings by Neighbourhood Group
plt.figure(figsize=(10,8))
sns.scatterplot(
    x="long", y="lat",
    hue="neighbourhood group",
    data=df, alpha=0.5, s=10, palette="Set1"
)
plt.title("Geographical Distribution of Listings by Neighbourhood Group")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Neighbourhood Group")
plt.show()
```

```python
# 8. Price Heatmap on Map (capped at $500 for visibility) 
plt.figure(figsize=(10,8))
sns.scatterplot(
    x="long", y="lat",
    hue=df["price"].clip(upper=500),
    data=df, alpha=0.6, s=10, palette="coolwarm"
)
plt.title("Geographical Price Distribution (Capped at $500)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar = plt.colorbar
plt.show()
```

```python
# 9. Reviews vs price scatter (sample for speed)
sample_df = airbnb_df[['price','number of reviews']].dropna().sample(n=min(5000, airbnb_df.shape[0]), random_state=42)
plt.figure(figsize=(6,4))
sns.scatterplot(data=sample_df, x='number of reviews', y='price', alpha=0.3, color='#1d3557')
plt.title('Price vs. Number of reviews (sample)')
plt.xlabel('Number of reviews')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

 This project analyzes hotel or Airbnb booking data to understand pricing trends, room availability, customer reviews, and host activities.

* Which types of rooms are most popular (entire home, private room, etc.)
